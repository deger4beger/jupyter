```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students_performance = pd.read_csv("D:\\python\\jupyter\\files\\StudentsPerformance.csv")
```

```python
students_performance.columns = [x.replace(" ", "_") for x in students_performance.columns]
```

```python
students_performance.math_score.hist()
```

```python
students_performance.plot.scatter(x="math_score", y="writing_score")
```

```python
ax = sns.lmplot(x="math_score", y="writing_score", hue="gender", data=students_performance, fit_reg=False)
ax.set_xlabels("Math score")
ax.set_ylabels("Writing score")
```
