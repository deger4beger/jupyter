```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://stepik.org/media/attachments/course/4852/iris.csv", index_col=0)
sns.set_style("whitegrid")
```

```python
df
```

```python
df.columns = [x.replace(" ", "_") for x in df.columns]
```

```python
sns.distplot(df.sepal_length)
```

```python
sns.distplot(df.sepal_width)
```

```python
sns.distplot(df.petal_length)
```

```python
sns.distplot(df.petal_width)
```

```python
sns.distplot(df.species)
```

```python
ax = sns.distplot( df["sepal_length"], color="skyblue", kde_kws={"label":"sepal_length"})
ax = sns.distplot( df["sepal_width"], color="red", kde_kws={"label":"sepal_width"})
ax.set_xlabel("sepal stats")
```

```python
sns.distplot( df["petal_length"] , color="purple", kde_kws={"label":"petal_length"})
sns.distplot( df["petal_width"] , color="black", kde_kws={"label":"petal_width"})
```

```python
ax = sns.violinplot(y=df["petal_length"], x=df["species"], palette="Set2")
```

```python
ax = sns.pairplot(df, kind="reg")
```
